```python
# Clean the dataset
df.isnull().sum()
```

```python
df['country'] = df['country'].fillna('unkown')
```

```python
df['director'] = df['director'].fillna('Not Available')
```

```python
df['cast'] = df['cast'].fillna('Not Available')
```

```python
df.duplicated().sum()
```

```python
# Data Preparation
df.drop_duplicates(inplace = True)
```

```python
# Remove leading/trailing spaces before converting
df['date_added'] = df['date_added'].astype(str).str.strip()

df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
df[['year_added','date_added']]
```

```python
# Movies Vs Tv Shows Analysis
df['type'].value_counts()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(data=df,x='type')

plt.title("Movies vs TV Shows")

plt.show()
```

Business Insight

Determine whether Netflix focuses more on movies or shows.

```python
# Content Growth Over Year
yearly_content = df.groupby('release_year').size()

plt.figure(figsize=(12,5))
yearly_content.plot()

plt.title("Netflix Content Growth")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.show()
```

Insight

Identify periods of rapid growth.

```python
%%writefile dashboard.py

import streamlit as st
import pandas as pd
import plotly.express as px

# Load Data
df = pd.read_csv("cleaned_netflix.csv")

# Title
st.title("Netflix Dashboard")

# Sidebar Filters
genre = st.sidebar.selectbox(
    "Genre",
    ["All"] + sorted(list(df['listed_in'].dropna().unique()))
)

year = st.sidebar.slider(
    "Release Year",
    int(df['release_year'].min()),
    int(df['release_year'].max()),
    int(df['release_year'].max())
)

# KPIs
st.metric(
    "Total Titles",
    len(df)
)

st.metric(
    "Movies",
    len(df[df['type'] == "Movie"])
)

st.metric(
    "TV Shows",
    len(df[df['type'] == "TV Show"])
)

# Movies vs TV Shows
fig = px.pie(
    df,
    names='type',
    title='Movies vs TV Shows'
)

st.plotly_chart(fig)

# Content Growth
growth = (
    df.groupby('release_year')
      .size()
      .reset_index(name='count')
)

fig = px.line(
    growth,
    x='release_year',
    y='count',
    title='Content Growth'
)

st.plotly_chart(fig)

# Top Countries
top_country = (
    df['country']
      .value_counts()
      .head(10)
)

fig = px.bar(
    x=top_country.index,
    y=top_country.values,
    labels={'x':'Country','y':'Titles'},
    title='Top 10 Countries'
)

st.plotly_chart(fig)
```
